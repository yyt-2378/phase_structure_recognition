```python

from ase.io import read
from PIL import Image, ImageDraw

cif_files ='F:/cv/new images/pyrdine/1.cif'
atoms_list = []

# 读取cif文件
atoms = read(cif_files)
    
# 将原子数据存储在列表中
atoms_list.append(atoms)

```

```python
# 输出各原子的实际坐标
for atom in atoms:
    symbol = atom.symbol
    position = atom.position
    print(f"{symbol} {position[0]:.6f} {position[1]:.6f} {position[2]:.6f}")
```

```python
for atom in atoms:
    position = atom.position
    position[1] = 0

    # 输出投影后的坐标
    print(f"{atom.symbol} {position[0]:.6f} {position[1]:.6f} {position[2]:.6f}")
```

```python


# 定义每种元素的灰度和半径
element_colors = {
    'Si': (130, 0.36875),
    'Al': (225, 0.36875),
    'O': (145, 0.228125),
    'H': (234, 0.1),
    'C': (136, 0.240625),
    'S': (246, 0.31875),
    'N': (192, 0.234375)
}


# 设置图像大小和缩放比例
image_size = (406, 271)
scale_xz = 4
scale_r = 6.00

# 创建一个白色背景图像
img = Image.new('L', image_size, color='white')

# 在xz平面上绘制原子
draw = ImageDraw.Draw(img)
for atom in atoms:
    element = atom.symbol
    x, y, z = atom.position
    if y == 0:
        # 计算绘制的圆形的位置和大小
        gray_value, radius = element_colors[element]
        x_pixel, z_pixel = int(x * scale_xz), int(z * scale_xz)
        radius_pixel = int(radius * scale_r)

        # 绘制圆形
        draw.ellipse((x_pixel - radius_pixel, z_pixel - radius_pixel,
                      x_pixel + radius_pixel, z_pixel + radius_pixel),
                     fill=gray_value)

# 保存图像
img.save('F:/cv/new images/pyrdine/1.png')
```
